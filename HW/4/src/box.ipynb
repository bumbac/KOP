```julia
include("simulated.jl")
using PlotlyJS
```

```julia
filename = "../data/100.dat"
instances = readFile(filename)
T = 40
cooling_factor = 0.999
inner_cycle = 200
frozen_limit = 2.7
```

```julia
ax = []
for a in 1:3
    profit, result, y, y_best = sa(instances[1], T, frozen_limit, inner_cycle, cooling_factor)
    push!(ax, scatter(y=y, mode="lines", opacity=0.8))
    push!(ax, scatter(y=y_best, mode="markers", opacity=0.8))
    println(profit)
end
```

```julia
plot([a for a in ax])
```
