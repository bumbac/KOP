```julia
function readSolution(filename::String)
    solution = []
    open(filename) do f
        while !eof(f)
            result = parse(Int64, readline(f))
            push!(solution, result)
        end
    end
    return solution
end
```

```julia
function writeSolution(filename::String, solution)
    open(filename, "w") do f
        for s in solution
            write(f, string(s)*"\n")
        end
    end
end
```
